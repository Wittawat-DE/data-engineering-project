**6.1 Create view to focus the data what we want to report.**

```python
CREATE VIEW `r2de-worksops.ws6_input.ws6_view`
AS
SELECT
  TIMESTAMP_MICROS(CAST(date / 1000 AS INT64)) AS date_converted
  ,thb_amount
  ,transaction_id
  ,product_id
  ,product_name
  ,quantity
  ,customer_id
  ,customer_name
  ,customer_country
FROM `r2de-worksops.ws6_input`
WHERE total_amount > 0;
```
